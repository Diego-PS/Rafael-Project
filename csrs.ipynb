```python
import os

directory = './amcr/csrs_ppp'

files = os.listdir(directory)
print(files)
```

```python
import csv

# open the file in the write mode
csv_f = open('csrs.csv', 'w', encoding='utf-8')

# create the csv writer
writer = csv.writer(csv_f)

# write a row to the csv file
writer.writerow(['tipo', 'data', 'duração', 'X', 'Y', 'Z', 'sigmaX', 'sigmaY', 'sigmaZ', 'IAR'])
```

```python
i = 0
for file in files:
    date = file[4:7]
    date = int(date)
    print(date)

    p1, p2 = file.split('_')
    duration, file_format = p2.split('.')
    duration = int(duration)
    print(duration)

    files[i] = (date, duration, file)
    i+=1

files.sort()
print(files)
```

```python
for file_tuple in files:
    date, duration, file = file_tuple
    f = open(directory+'/'+file, 'r')
    message = f.read()

    #print(message)

    f.close()
    lines = message.split('\n')

    #print(lines)

    lines = [line.split() for line in lines]

    IAR_line = []
    for line in lines:
        if len(line) > 0:
            if line[0] == 'IAR':
                IAR_line = line

    POS_lines = []
    for line in lines:
        if len(line) > 0:
            if line[0] == 'POS':
                POS_lines.append(line)

    X_line = []
    for line in POS_lines:
        if (line[1] == 'X'):
            X_line = line

    Y_line = []
    for line in POS_lines:
        if (line[1] == 'Y'):
            Y_line = line

    Z_line = []
    for line in POS_lines:
        if (line[1] == 'Z'):
            Z_line = line

    csv_row = ['csrs', str(date), str(duration), X_line[5], Y_line[5], Z_line[5], X_line[6], Y_line[6], Z_line[6], IAR_line[1]]
    writer.writerow(csv_row)
    
```
